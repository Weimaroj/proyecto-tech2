**Oferta-Producción**

```python
# Patrones Oferta - Producción

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Convertir la columna de año a numérico
df_oferta_energia_col['Años__ESTANDAR'] = pd.to_numeric(df_oferta_energia_col['Años__ESTANDAR'], errors='coerce')
df_produccion_energia_col['Años__ESTANDAR'] = pd.to_numeric(df_produccion_energia_col['Años__ESTANDAR'], errors='coerce')
df_oferta_energia_latam['Años__ESTANDAR'] = pd.to_numeric(df_oferta_energia_latam['Años__ESTANDAR'], errors='coerce')
df_produccion_energia_latam['Años__ESTANDAR'] = pd.to_numeric(df_produccion_energia_latam['Años__ESTANDAR'], errors='coerce')

# Agregar la oferta y producción por año y tipo de energía
off_col = df_oferta_energia_col.groupby(['Años__ESTANDAR', 'Tipo de energía__Primaria_y_Secundaria'])['value'].sum().unstack()
prod_col = df_produccion_energia_col.groupby(['Años__ESTANDAR', 'Tipo de energía__Primaria_Secundaria'])['value'].sum().unstack()
off_latam = df_oferta_energia_latam.groupby(['Años__ESTANDAR', 'Tipo de energía__Primaria_y_Secundaria'])['value'].sum().unstack()
prod_latam = df_produccion_energia_latam.groupby(['Años__ESTANDAR', 'Tipo de energía__Primaria_Secundaria'])['value'].sum().unstack()

# Visualizar la oferta y producción de energía en Colombia
plt.figure(figsize=(12, 6))
plt.plot(off_col.index, off_col.sum(axis=1), label='Oferta Colombia', marker='o')
plt.plot(prod_col.index, prod_col.sum(axis=1), label='Producción Colombia', marker='s')
plt.xlabel('Año')
plt.ylabel('Miles de barriles equivalentes de petróleo')
plt.title('Evolución de la Oferta y Producción de Energía en Colombia')
plt.legend()
plt.grid(True)
plt.show()

# Visualizar la oferta y producción de energía en LATAM
plt.figure(figsize=(12, 6))
plt.plot(off_latam.index, off_latam.sum(axis=1), label='Oferta LATAM', marker='o')
plt.plot(prod_latam.index, prod_latam.sum(axis=1), label='Producción LATAM', marker='s')
plt.xlabel('Año')
plt.ylabel('Miles de barriles equivalentes de petróleo')
plt.title('Evolución de la Oferta y Producción de Energía en LATAM')
plt.legend()
plt.grid(True)
plt.show()

# Correlación entre oferta y producción en Colombia y LATAM
corr_col = off_col.corrwith(prod_col)
corr_latam = off_latam.corrwith(prod_latam)
print("Correlación entre oferta y producción por tipo de energía en Colombia:\n", corr_col)
print("\nCorrelación entre oferta y producción por tipo de energía en LATAM:\n", corr_latam)
```

**Consumo, oferta y producción**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Fusionar bases de datos para análisis conjunto
df_colombia = df_consumo_energia_col.merge(df_oferta_energia_col, on=["Años__ESTANDAR", "Grupo_Energía"], suffixes=("_consumo", "_oferta"))
df_colombia = df_colombia.merge(df_produccion_energia_col, on=["Años__ESTANDAR", "Grupo_Energía"], suffixes=("", "_produccion"))

df_latam = df_consumo_energia_latam.merge(df_oferta_energia_latam, on=["Años__ESTANDAR", "Grupo_Energía"], suffixes=("_consumo", "_oferta"))
df_latam = df_latam.merge(df_produccion_energia_latam, on=["Años__ESTANDAR", "Grupo_Energía"], suffixes=("", "_produccion"))

# Calcular correlaciones entre producción, oferta y consumo
print("\nCorrelaciones Colombia:\n", df_colombia[["value_consumo", "value_oferta", "value"]].corr())
print("\nCorrelaciones LATAM:\n", df_latam[["value_consumo", "value_oferta", "value"]].corr())

# Visualizar tendencias de producción, oferta y consumo en el tiempo
plt.figure(figsize=(12, 5))
sns.lineplot(x="Años__ESTANDAR", y="value_consumo", data=df_colombia, label="Consumo", marker="o")
sns.lineplot(x="Años__ESTANDAR", y="value_oferta", data=df_colombia, label="Oferta", marker="s")
sns.lineplot(x="Años__ESTANDAR", y="value", data=df_colombia, label="Producción", marker="d")
plt.title("Evolución de Consumo, Oferta y Producción en Colombia")
plt.ylabel("Energía (Miles de barriles equivalentes de pétroleo)")
plt.xlabel("Año")
plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(12, 5))
sns.lineplot(x="Años__ESTANDAR", y="value_consumo", data=df_latam, label="Consumo", marker="o")
sns.lineplot(x="Años__ESTANDAR", y="value_oferta", data=df_latam, label="Oferta", marker="s")
sns.lineplot(x="Años__ESTANDAR", y="value", data=df_latam, label="Producción", marker="d")
plt.title("Evolución de Consumo, Oferta y Producción en LATAM")
plt.ylabel("Energía (Miles de barriles equivalentes de pétroleo)")
plt.xlabel("Año")
plt.legend()
plt.grid()
plt.show()

# Calcular diferencias entre producción y consumo para detectar déficit o superávit
df_colombia["Balance_Energetico"] = df_colombia["value"] - df_colombia["value_consumo"]
df_latam["Balance_Energetico"] = df_latam["value"] - df_latam["value_consumo"]

plt.figure(figsize=(12, 5))
sns.lineplot(x="Años__ESTANDAR", y="Balance_Energetico", data=df_colombia, label="Balance Colombia", marker="o")
sns.lineplot(x="Años__ESTANDAR", y="Balance_Energetico", data=df_latam, label="Balance LATAM", marker="s")
plt.axhline(0, color="black", linestyle="--", alpha=0.7)
plt.title("Balance Energético: Diferencia entre Producción y Consumo")
plt.ylabel("Excedente/Déficit Energético")
plt.xlabel("Año")
plt.legend()
plt.grid()
plt.show()
```

**Consumo de energía eléctrica - Consumo de energía total**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Fusionar datasets por Año para comparar consumo total y consumo eléctrico
df_colombia = df_consumo_energia_col.groupby("Años__ESTANDAR")["value"].sum().reset_index().rename(columns={"value": "Consumo_Total"})
df_electrica_colombia = df_consumo_energia_electrica_col.groupby("Años__ESTANDAR")["value"].sum().reset_index().rename(columns={"value": "Consumo_Electrico"})

df_latam = df_consumo_energia_latam.groupby("Años__ESTANDAR")["value"].sum().reset_index().rename(columns={"value": "Consumo_Total"})
df_electrica_latam = df_consumo_energia_electrica_latam.groupby("Años__ESTANDAR")["value"].sum().reset_index().rename(columns={"value": "Consumo_Electrico"})

df_colombia = df_colombia.merge(df_electrica_colombia, on="Años__ESTANDAR")
df_latam = df_latam.merge(df_electrica_latam, on="Años__ESTANDAR")

# Calcular el porcentaje de energía eléctrica sobre el total
df_colombia["Porcentaje_Electrico"] = (df_colombia["Consumo_Electrico"] / df_colombia["Consumo_Total"]) * 100
df_latam["Porcentaje_Electrico"] = (df_latam["Consumo_Electrico"] / df_latam["Consumo_Total"]) * 100

# Mostrar correlación entre consumo total y consumo eléctrico
print("\nCorrelación Colombia:\n", df_colombia[["Consumo_Total", "Consumo_Electrico"]].corr())
print("\nCorrelación LATAM:\n", df_latam[["Consumo_Total", "Consumo_Electrico"]].corr())

# Visualización del consumo total vs consumo eléctrico en el tiempo
plt.figure(figsize=(12, 5))
sns.lineplot(x="Años__ESTANDAR", y="Consumo_Total", data=df_colombia, label="Consumo Total", marker="o")
sns.lineplot(x="Años__ESTANDAR", y="Consumo_Electrico", data=df_colombia, label="Consumo Eléctrico", marker="s")
plt.title("Evolución del Consumo Total y Eléctrico en Colombia")
plt.ylabel("Consumo de Energía (TWh o unidad correspondiente)")
plt.xlabel("Año")
plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(12, 5))
sns.lineplot(x="Años__ESTANDAR", y="Consumo_Total", data=df_latam, label="Consumo Total", marker="o")
sns.lineplot(x="Años__ESTANDAR", y="Consumo_Electrico", data=df_latam, label="Consumo Eléctrico", marker="s")
plt.title("Evolución del Consumo Total y Eléctrico en LATAM")
plt.ylabel("Consumo de Energía (TWh o unidad correspondiente)")
plt.xlabel("Año")
plt.legend()
plt.grid()
plt.show()

# Visualizar la proporción del consumo eléctrico sobre el total
plt.figure(figsize=(12, 5))
sns.lineplot(x="Años__ESTANDAR", y="Porcentaje_Electrico", data=df_colombia, label="Colombia", marker="o")
sns.lineplot(x="Años__ESTANDAR", y="Porcentaje_Electrico", data=df_latam, label="LATAM", marker="s")
plt.axhline(50, color="red", linestyle="--", alpha=0.7, label="50% referencia")
plt.title("Porcentaje de Energía Eléctrica sobre el Consumo Total")
plt.ylabel("Porcentaje (%)")
plt.xlabel("Año")
plt.legend()
plt.grid()
plt.show()
```
